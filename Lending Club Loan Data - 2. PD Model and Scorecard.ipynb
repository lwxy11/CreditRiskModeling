## Build a Logistic Regression Model with P-values

### Build class with p values

```python
from sklearn import linear_model
import scipy.stats as stat

class LogisticRegression_with_p_values:
    
    # an instance of this class initialize is by inheriting from the original LogisticRegression class
    def __init__(self,*args,**kwargs): 
        self.model = linear_model.LogisticRegression(*args, **kwargs)
        
    # define a fit function which overrides the original fit available in the LogisticRegression class
    def fit(self,X,y):
        self.model.fit(X,y)
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom, (X.shape[1],1)).T
        F_ij = np.dot((X / denom).T,X)
        Cramer_Rao = np.linalg.inv(F_ij)
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.model.coef_[0] / sigma_estimates
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores]
        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_
        self.p_values = p_values
```

```python
reg = LogisticRegression_with_p_values()
```

```python
reg.fit(inputs_train, y_train)
```

```python
summary_table = pd.DataFrame(columns = ['Feature name'], data = feature_name)
summary_table['Coefficients'] = np.transpose(reg.coef_)
# add intercept
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
# add p-values
p_values = reg.p_values
p_values = np.append(np.nan, np.array(p_values))
summary_table['p_values'] = p_values
summary_table
```

If a p-value is lower than 0.05, we conclude the coefficient of a variable is significant.

## Setting Cut-offs

```python
df_cutoffs = pd.concat([pd.DataFrame(thresholds),
                        pd.DataFrame(fpr),
                        pd.DataFrame(tpr)], axis=1)
df_cutoffs.columns = ['thresholds','fpr','tpr']
df_cutoffs.head()
```

```python
df_cutoffs['thresholds'][0] = 1 - 1/np.power(10,16)
df_cutoffs.head()
```

```python
df_cutoffs['Score'] = ((np.log(df_cutoffs['thresholds']/(1-df_cutoffs['thresholds'])) - min_sum_coef) * \
                       ((max_score-min_score)/(max_sum_coef-min_sum_coef)) + min_score).round()
df_cutoffs.head()
```

```python
df_cutoffs['Score'][0] = max_score
df_cutoffs.head()
```

```python
df_cutoffs.tail()
```

```python
def n_approved(p):
    return np.where(df_actual_predicted_probs['y_hat_test_proba'] >= p,
                       1, 0).sum()
```

```python
df_cutoffs['N Approved'] = df_cutoffs['thresholds'].apply(n_approved)
df_cutoffs['N Rejected'] = df_actual_predicted_probs['y_hat_test_proba'].shape[0] - df_cutoffs['N Approved']
df_cutoffs['Approval Rate'] = df_cutoffs['N Approved'] / df_actual_predicted_probs['y_hat_test_proba'].shape[0]
df_cutoffs['Rejection Rate'] = 1 - df_cutoffs['Approval Rate'] 
```
